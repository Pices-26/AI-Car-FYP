```python
#imports
from skimage.io import imread
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read image
img = imread('oldData/test/straight.jpg')
plt.imshow(img)
```

```python
#image processing function
def preprocessImg(img):
    img = img[70:240,0:432]
    h,w,d = img.shape
    img = cv2.resize(img,(int(w/3),int(h/3)))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
    img = cv2.GaussianBlur(img, (3, 3), 0)
    img = img/255.0
    return img
```

```python
#return image + show
new_img = preprocessImg(img)
plt.imshow(new_img)
new_img.shape
```
